```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Import iris dataset


```python
df=pd.read_csv('/content/8836201/iris.csv')
```

displaying first 5 record

Displaying information about the data set

```python
df.info()
```

Counting number of varieties

```python
df["variety"].value_counts()
```

checking for duplicates

```python
df[df.duplicated()]
```

Visualisation

Histogram

```python
df['sepal.length'].plot(kind='hist')
plt.show()
```

```python
df.hist()
plt.show()
```

box plot

```python
df.boxplot()
plt.show()
```

```python
df.boxplot(by='variety')
plt.show()
```

scatter plot

```python
df.plot(kind='scatter',x='sepal.width',y='petal.width',s=15)
plt.show()
```

```python
df.plot(kind='scatter',x='sepal.length',y='petal.length',s=15,color='G')
plt.show()
```

```python
sns.scatterplot(data=df,x="petal.length",y="petal.width",hue="variety")
plt.show
```

distribution chart

```python
df.plot()
plt.title("All graph")
plt.show()
```

scatter matrix

```python
sns.pairplot(df,hue='variety')
plt.title("")
plt.show()
```

Bubble chart

```python
fig = plt.figure(figsize = (14, 8))
ax = sns.scatterplot(data=df, x="sepal.length", y="sepal.width",legend=False, sizes=(20, 2000))
ax.set(title='Bubble Chart - Seaborn')
plt.show()
```

andrews curve

```python
from pandas.plotting import andrews_curves
p=andrews_curves(df,'variety')
```

density chart

```python
df.plot(kind='density')
plt.show()
```

parallel chart

```python
pd.plotting.parallel_coordinates(df.drop("sepal.width",axis=1),"variety",colormap="cool")
plt.show()
```
