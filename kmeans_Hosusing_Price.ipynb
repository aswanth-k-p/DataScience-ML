```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
diff=pd.read_csv('/content/Housing_Price.csv')
```

```python
x1=diff.loc[:,['longitude','latitude']].values
model=KMeans(n_clusters=4, init='k-means++' ,max_iter=500, random_state=42).fit(x1)
```

```python
centroids=model.cluster_centers_
y_KMeans=model.fit_predict(x1)
plt.figure(figsize=(20,10))
plt.scatter(x1[y_KMeans==0,0],x1[y_KMeans==0,1],s=100,c='grey',label='cluster1')
plt.scatter(x1[y_KMeans==1,0],x1[y_KMeans==1,1],s=100,c='blue',label='cluster2')
plt.scatter(x1[y_KMeans==2,0],x1[y_KMeans==2,1],s=100,c='purple',label='cluster3')
plt.scatter(x1[y_KMeans==3,0],x1[y_KMeans==3,1],s=100,c='black',label='cluster4')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.legend()
plt.show()
```
